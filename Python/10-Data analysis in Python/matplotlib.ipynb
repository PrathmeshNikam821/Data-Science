##### line plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y)
# plt.plot(y,x)
plt.xlabel('X axis ')
plt.ylabel('Y axis')
plt.title('Basic line plot')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

#create a customized line plot

plt.plot(x,y,color='red',linestyle='-.',marker='o' ,linewidth='3')
plt.grid(True)
```

```python
#multiple data 

x =[ 1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,8,27,81,125]
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.plot(x,y1,color='orange',linestyle='-.')
plt.title('plot 1')

plt.subplot(2,2,2)
plt.plot(y1,y2,color='red',marker='o')
plt.title('plot 2')

plt.subplot(2,2,3)
plt.plot(y1,y2,color='blue',marker='o')
plt.title('plot 3')

plt.subplot(2,2,4)
plt.plot(y1,y2,color='pink',marker='1')
plt.title('plot 4')
```
