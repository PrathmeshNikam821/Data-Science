##### line plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y)
# plt.plot(y,x)
plt.xlabel('X axis ')
plt.ylabel('Y axis')
plt.title('Basic line plot')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

#create a customized line plot

plt.plot(x,y,color='red',linestyle='-.',marker='o' ,linewidth='3')
plt.grid(True)
```

```python
#multiple data 

x =[ 1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,8,27,81,125]
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.plot(x,y1,color='orange',linestyle='-.')
plt.title('plot 1')

plt.subplot(2,2,2)
plt.plot(y1,y2,color='red',marker='o')
plt.title('plot 2')

plt.subplot(2,2,3)
plt.plot(y1,y2,color='blue',marker='o')
plt.title('plot 3')

plt.subplot(2,2,4)
plt.plot(y1,y2,color='pink',marker='1')
plt.title('plot 4')
```

#### Bar graph


```python
categories = ['a','b','c','d','e']
values=[1,2,3,4,5]

#create a bar plot

plt.bar(categories,values ,color='orange')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('categories vs values ')
plt.show()
```

#### Histograms 

uesd to represent the distribution of a dataset.


divide the dataset into bins and count the number of data points in each bin.

```python
#sample data 


data = [1,1,1,3,4,5,2,2,2,2,2,2,4,5,4,5,5,5,3,3,3]

#create a histogram 
plt.hist(data,bins=5,color='orange',edgecolor='black')

```

##### scatterplot

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

#create a scatterplot

plt.scatter(x,y,color='black',marker='x')
```

#### pie chart

```python
labels =['A','B','C','D']

sizes=[30,40,20,10]

colors=['gold','yellowgreen','lightcoral','lightskyblue']

explode = (0.2,0.2,0.2,0.2)

plt.pie(sizes,labels=labels,colors=colors ,autopct="%1.1f%%",shadow=True , explode=explode)
```

```python

#sales data visualization 

import pandas as pd 
sales_data_df= pd.read_csv('sales_data.csv')

sales_data_df.head()
```

```python
sales_data_df.info()
```

```python
#plot total sales by product
total_sales_by_product = sales_data_df.groupby('Product Category')['Total Revenue'].sum()

```

```python
total_sales_by_product.plot(kind='bar',color='teal')
```

```python
#plot sales trend over time 

sales_trend = sales_data_df.groupby('Date')['Total Revenue'].sum().reset_index()
sales_trend

plt.plot(sales_trend['Date'], sales_trend['Total Revenue'])
```
